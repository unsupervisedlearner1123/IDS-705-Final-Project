```python
import pandas as pd
import boto3

BUCKET_NAME = 'aws_bucket_name' # replace with your bucket name

key = "train.csv"

df_full = pd.read_csv(
    f"s3://{BUCKET_NAME}/{key}",
    storage_options={
        "key": 'AWS_SECRET_KEY',
        "secret": 'AWS_SECRET_TOKEN',
        # "token": AWS_SESSION_TOKEN,
    }, iterator = True, chunksize=100000
)

ls = []
for chunk in df_full:
    ls.append(chunk)
    print("Chunk finished.")
    pass

df = pd.concat(ls)
```
